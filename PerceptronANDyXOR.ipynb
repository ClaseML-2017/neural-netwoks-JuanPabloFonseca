```python
import pandas as pd
import numpy as np
from scipy import sparse
from sys import maxint
from math import pow
from math import fabs
from __future__ import division
from scipy.stats import norm
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import mean_squared_error
from sklearn import linear_model
from sklearn import preprocessing
import matplotlib.pyplot as plt
from random import uniform
```

```python
# función lógica AND
datosXand = [[0, 0], [0, 1], [1, 0], [1, 1]]
datosYand = [0, 0, 0, 1]
p = 2
```

```python
def g(x, coeff):
    ret = 0
    for i in range(0,p+1):
        ret += coeff[i]*x[i]
    if ret >= 0:
        ret = 1
    else:
        ret = 0
    return ret
```

```python
def eq(x, coeff):
    eq = 0
    for i in range(0,p+1):
        eq += coeff[i]*x[i]
    return eq
```

## AND

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Axes3D.scatter(ax,[x[0] for x in datosXand],[x[1] for x in datosXand], datosYand)
plt.show()
```

```python
w = [1 for i in range(0,p+1)]
eta = 0.01
for k in range (0,1000):
    for i in range (0,len(datosXand)):
        xi0 = [1]
        xidemas=[datosXand[i][j] for j in range(0,p)]
        xi = xi0 + xidemas
        yi = datosYand[i]
        error_actual = yi - g(xi,w)
        for j in range(len(w)):
            w[j] = w[j] + eta * error_actual * xi[j]
plt.scatter([x[0] for x in datosXand],[x[1] for x in datosXand])
x1 = np.linspace(0, 1, 200, endpoint=False)
plt.plot(x1,-(w[0]/w[2])-(w[1]/w[2])*x1)
plt.show()
print "Sí divide bien; hasta la derecha la recta está {0} arriba del punto, como debe ser".format(-(w[0]/w[2])-(w[1]/w[2])*x1[-1])
```
