```python
import tensorflow as tf
from __future__ import division
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# variables
lim = 1.2
densidad = 3000
a = np.random.uniform(-lim,lim,densidad)
b = np.random.uniform(-lim,lim,densidad)
f = a * a
d = b * b
y = f + d <= 1
y = y.astype(int)
X=np.asarray([[a[k],b[k],f[k],d[k]] for k in range(0,len(a))])
Y=np.asarray([[elem] for elem in y])
```

```python
# red neuronal
input_size=4
middle_layer=4
hidden_layers=4
output_size=1
x=tf.placeholder(tf.float32,shape=[None,input_size])
y_=tf.placeholder(tf.float32,shape=[None,output_size])
W_layer1 = tf.Variable(tf.random_uniform([input_size,middle_layer], -1, 1), name="W_layer1")
b_layer1 = tf.Variable(tf.zeros([middle_layer]), name="b_layer1")
W_layer2 = tf.Variable(tf.random_uniform([hidden_layers,output_size], -1, 1), name="W_layer2")
b_layer2 = tf.Variable(tf.zeros([output_size]), name="b_layer2")
```

```python
hidden_1=tf.nn.sigmoid(tf.matmul(x,W_layer1)+b_layer1)
y= tf.nn.sigmoid(tf.matmul(hidden_1,W_layer2)+b_layer2)
lossfn= tf.reduce_mean(( (y_ * tf.log(y)) + ((1 - y_) * tf.log(1.0 - y)) ) * -1)
train_step = tf.train.GradientDescentOptimizer(0.01).minimize(lossfn)
```

```python
# iniciar la sesiÃ³n
init = tf.global_variables_initializer()
sess = tf.Session()
```

```python
sess.run(init)
for i in range(6000):
  sess.run(train_step, feed_dict={x: X, y_: Y})
```

### Error

```python
predCorrecta = tf.equal(tf.round(y),y_) 
acc = tf.reduce_mean(tf.cast(predCorrecta, tf.float32))
print(sess.run(acc, feed_dict={x: X, y_: Y}))
```

```python
# graficar
pred = sess.run(y,feed_dict={x:X})
plt.scatter(a, b, c=['orange' if i < 0.5 else 'blue' for i in pred])
plt.ylim([-1.1,1.1])
plt.xlim([-1.3,1.3])
plt.show()
```

#### Se observa que la red neuronal sÃ­ rescata la forma del cÃ­rculo.
